```python
import matplotlib.pyplot as plt 
import numpy as np 
import math 
  
x = np.linspace(-10, 10, 100) 
z = 1/(1 + np.exp(-x)) 
  
plt.plot(x, z) 
plt.xlabel("x") 
plt.ylabel("Sigmoid(X)") 
  
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def reLu(x):
  return np.array([xi if xi >= 0 else 0 for xi in x])
x=np.arange(-3,2,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, reLu(x))
    
```

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x, lam=1):
  return (np.exp(lam*x)-np.exp(-lam*x))/(np.exp(lam*x)+np.exp(-lam*x))

x=np.arange(-10,10,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, tanh(x))
```

```python
import numpy as np
import matplotlib.pyplot as plt
def logsigmoid(x, lam=1):
  return np.log(1/(1+np.exp(-lam*x)))

x=np.arange(-10,10,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, logsigmoid(x))
```
