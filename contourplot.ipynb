```
import matplotlib.pyplot as plt
import numpy as np
```

```
x1 = np.linspace(-4.0, 5.0, 100)
x2 = np.linspace(-4.0, 5.0, 100)
```

```
xx1, xx2 = np.meshgrid(x1,x2)
```

```
# Quadratic Function

fig, ax = plt.subplots(1,2)
fig.set_size_inches(15, 5)
fig.colorbar(ax[0].contourf(xx1,xx2, xx1**2+xx2**2 -10))
fig.colorbar(ax[1].contour(xx1,xx2, xx1**2+xx2**2 -10))
```

```
# Cubic Function

fig, ax = plt.subplots(1,2)
fig.set_size_inches(15, 5)
fig.colorbar(ax[0].contourf(xx1,xx2, xx1**3+xx2**3 -10))
fig.colorbar(ax[1].contour(xx1,xx2, xx1**3+xx2**3 -10))
```
